```python
def get_tier(followers: int) -> str:
    if followers < 1000:
        return "Seeder"
    elif followers < 10000:
        return "Nano"
    elif followers < 100000:
        return "Micro"
    elif followers < 1000000:
        return "Macro"
    else:
        return "Celebrity"


df_rn['tier']=df_rn['followers'].apply(get_tier)
```

```python
#add unique id
import uuid
def short_uuid(prefix="t"):
    return f"{prefix}{uuid.uuid4().hex[:8]}"
df_rn["uniqueid"]=[short_uuid("r") for _ in range(len(df_rn))]
```
